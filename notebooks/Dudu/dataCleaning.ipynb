# Import dos DF

```python
import pandas as pd

df_Treino = pd.read_csv("initTraining.csv")
df_Teste = pd.read_csv("initTest.csv")

df_Treino
```

```python
print(df_Treino.info())
```

```python
print(df_Treino.describe())
```

# ComeÃ§ando a Limpeza de Colunas

```python
df_Treino = df_Treino.loc[:, ~df_Treino.columns.str.contains('^Unnamed')]
df_Teste = df_Teste.loc[:, ~df_Teste.columns.str.contains('^Unnamed')]
```

```python
print(df_Teste.isnull().sum())
```

```python
limite_nulos = len(df_Treino) * 0.2
df_Treino = df_Treino.dropna(thresh=limite_nulos, axis=1)

limite_nulos = len(df_Teste) * 0.2
df_Teste = df_Teste.dropna(thresh=limite_nulos, axis=1)
```

```python
df_Treino['ts_primeiro_acesso'] = pd.to_datetime(df_Treino['ts_primeiro_acesso'], unit='s', errors='coerce')
df_Treino['ts_ultimo_acesso'] = pd.to_datetime(df_Treino['ts_ultimo_acesso'], unit='s', errors='coerce')

df_Teste['ts_primeiro_acesso'] = pd.to_datetime(df_Teste['ts_primeiro_acesso'], unit='s', errors='coerce')
df_Teste['ts_ultimo_acesso'] = pd.to_datetime(df_Teste['ts_ultimo_acesso'], unit='s', errors='coerce')
```

```python
df_Treino = df_Treino.rename(columns={
    'ts_primeiro_acesso': 'dt_primeiro_acesso',
    'ts_ultimo_acesso': 'dt_ultimo_acesso'
})

df_Teste = df_Teste.rename(columns={
    'ts_primeiro_acesso': 'dt_primeiro_acesso',
    'ts_ultimo_acesso': 'dt_ultimo_acesso'
})
```

# Valores Faltantes

```python
print(df_Treino.isnull().sum())
```

# Null Treatment

```python
# lista de colunas por tipo
dt_cols  = df_Treino.select_dtypes(include=['datetime64[ns]', 'datetime64[ns, UTC]']).columns.tolist()
num_cols = df_Treino.select_dtypes(include=['number']).columns.tolist()
obj_cols = df_Treino.select_dtypes(include=['object', 'category']).columns.tolist()

print("Datetime:",  dt_cols)
print("NumÃ©ricas:", num_cols)
print("CategÃ³ricas:", obj_cols)
```

```python
print("===== DATETIME =====")
print(df_Treino[dt_cols].isna().sum(), end="\n\n")

print("===== NUMÃ‰RICAS =====")
print(df_Treino[num_cols].isna().sum(), end="\n\n")

print("===== CATEGÃ“RICAS =====")
print(df_Treino[obj_cols].isna().sum())
```

```python
missing_pct = df_Treino[num_cols].isna().mean() * 100
print(missing_pct.sort_values(ascending=False))
```

## Drop >70%

```python
to_drop = [
    'vl_medio_tempo_questionario_avaliado',
    'vl_medio_tempo_questionario',
    'vl_desempenho_usuario',
    'vl_media_notas'
]
df_Treino = df_Treino.drop(columns=to_drop)
df_Teste = df_Teste.drop(columns=to_drop)
```

## Input ~47% MÃ©dia

```python
to_impute = ['nr_intervalos_uso', 'vl_medio_atividade_diaria']
medians_trn = df_Treino[to_impute].median()
medians_tst = df_Teste[to_impute].median()

df_Treino[to_impute] = df_Treino[to_impute].fillna(medians_trn)
df_Teste[to_impute] = df_Teste[to_impute].fillna(medians_tst)
```

```python
print(df_Treino[to_impute].isna().sum())
print(df_Teste[to_impute].isna().sum())
```

# Limpeza de Duplicatas

```python
print(f"Antes: {len(df_Treino)} linhas")
print(f"Antes: {len(df_Teste)} linhas\n")

print("Duplicatas em treino:", df_Treino.duplicated().sum())
print("Duplicatas em teste:", df_Teste.duplicated().sum())

df_Treino = df_Treino.drop_duplicates(keep='last')
df_Teste = df_Teste.drop_duplicates(keep='last')

print(f"\nDepois: {len(df_Treino)} linhas")
print(f"Depois: {len(df_Teste)} linhas\n")

print("ApÃ³s remoÃ§Ã£o â€” treino:", df_Treino.duplicated().sum())
print("ApÃ³s remoÃ§Ã£o â€” teste:", df_Teste.duplicated().sum())
```

# Ajuste de Datas InvÃ¡lidas

```python
print("df_Treino - dt_primeiro_acesso == 1970:", (df_Treino['dt_primeiro_acesso'].dt.year == 1970).sum())
print("df_Treino - dt_ultimo_acesso   == 1970:", (df_Treino['dt_ultimo_acesso'].dt.year == 1970).sum())

print("df_Teste - dt_primeiro_acesso == 1970:", (df_Teste['dt_primeiro_acesso'].dt.year == 1970).sum())
print("df_Teste - dt_ultimo_acesso   == 1970:", (df_Teste['dt_ultimo_acesso'].dt.year == 1970).sum())

df_Treino = df_Treino[df_Treino['dt_primeiro_acesso'].dt.year != 1970]
df_Treino = df_Treino[df_Treino['dt_ultimo_acesso'].dt.year != 1970]

df_Teste = df_Teste[df_Teste['dt_primeiro_acesso'].dt.year != 1970]
df_Teste = df_Teste[df_Teste['dt_ultimo_acesso'].dt.year != 1970]
```

# MÃ©todo de verificaÃ§Ã£o de abandono - 90d+ de AusÃªncia

```python
meses_ferias = [1, 2, 7, 12]

def verificar_abandono(row):
    ultimo_acesso = row['dt_ultimo_acesso']
    dias_sem_acesso = row['nr_dias_desde_ultimo_acesso']

    if ultimo_acesso.month in meses_ferias:
        return 0
    elif dias_sem_acesso > 90:
        return 1
    else:
        return 0
    
df_Treino['abandono'] = df_Treino.apply(verificar_abandono, axis=1)
df_Teste['abandono'] = df_Teste.apply(verificar_abandono, axis=1)
```

# RemoÃ§Ã£o das colunas desnecessÃ¡rias

```python
colunas_remover = ['dt_primeiro_acesso', 'dt_ultimo_acesso', 'ds_cidade_usuario']
df_Treino = df_Treino.drop(columns=colunas_remover)
df_Teste = df_Teste.drop(columns=colunas_remover)
```
