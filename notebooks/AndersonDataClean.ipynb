# Removendo colunas não relevantes

```python
df_trn = df_trn.loc[:, ~df_trn.columns.str.contains('^Unnamed')]
df_tst = df_tst.loc[:, ~df_tst.columns.str.contains('^Unnamed')]
```

```python
print(df_tst.isnull().sum())
```

```python
limite_nulos = len(df_trn) * 0.2
df_trn = df_trn.dropna(thresh=limite_nulos, axis=1)

limite_nulos = len(df_tst) * 0.2
df_tst = df_tst.dropna(thresh=limite_nulos, axis=1)
```

# Convertendo timeStamps

```python
df_trn['ts_primeiro_acesso'] = pd.to_datetime(df_trn['ts_primeiro_acesso'], unit='s', errors='coerce')
df_trn['ts_ultimo_acesso'] = pd.to_datetime(df_trn['ts_ultimo_acesso'], unit='s', errors='coerce')

df_tst['ts_primeiro_acesso'] = pd.to_datetime(df_tst['ts_primeiro_acesso'], unit='s', errors='coerce')
df_tst['ts_ultimo_acesso'] = pd.to_datetime(df_tst['ts_ultimo_acesso'], unit='s', errors='coerce')
```

```python
df_trn = df_trn.rename(columns={
    'ts_primeiro_acesso': 'dt_primeiro_acesso',
    'ts_ultimo_acesso': 'dt_ultimo_acesso'
})

df_tst = df_tst.rename(columns={
    'ts_primeiro_acesso': 'dt_primeiro_acesso',
    'ts_ultimo_acesso': 'dt_ultimo_acesso'
})
```

### Drop das colunas com >70% de missing

```python
to_drop = [
    'vl_medio_tempo_questionario_avaliado',
    'vl_medio_tempo_questionario',
    'vl_desempenho_usuario',
    'vl_media_notas'
]
df_trn = df_trn.drop(columns=to_drop)
df_tst = df_tst.drop(columns=to_drop)
```

### Imputar colunas com ~47% de missing pela mediana

```python
to_impute = ['nr_intervalos_uso', 'vl_medio_atividade_diaria']
medians_trn = df_trn[to_impute].median()
medians_tst = df_tst[to_impute].median()

df_trn[to_impute] = df_trn[to_impute].fillna(medians_trn)
df_tst[to_impute] = df_tst[to_impute].fillna(medians_tst)
```

### Verificando que não sobrou nenhum NaN

```python
print(df_trn[to_impute].isna().sum())
print(df_tst[to_impute].isna().sum())
```

# Ajustando data invalidadas

### Verificando quantas linhas nos datasets tem valores invalidos (1970 é o ponto inicial das bibliotecas)

```python
print("df_trn - dt_primeiro_acesso == 1970:", (df_trn['dt_primeiro_acesso'].dt.year == 1970).sum())
print("df_trn - dt_ultimo_acesso   == 1970:", (df_trn['dt_ultimo_acesso'].dt.year == 1970).sum())

print("df_tst - dt_primeiro_acesso == 1970:", (df_tst['dt_primeiro_acesso'].dt.year == 1970).sum())
print("df_tst - dt_ultimo_acesso   == 1970:", (df_tst['dt_ultimo_acesso'].dt.year == 1970).sum())
```

```python
df_trn = df_trn[df_trn['dt_primeiro_acesso'].dt.year != 1970]
df_trn = df_trn[df_trn['dt_ultimo_acesso'].dt.year != 1970]

df_tst = df_tst[df_tst['dt_primeiro_acesso'].dt.year != 1970]
df_tst = df_tst[df_tst['dt_ultimo_acesso'].dt.year != 1970]
```

# Supondo que alunos que não entram a mais de 90 dias fora do periodo de ferias abandonaram

```python
meses_ferias = [1, 2, 7, 12]

def verificar_abandono(row):
    ultimo_acesso = row['dt_ultimo_acesso']
    dias_sem_acesso = row['nr_dias_desde_ultimo_acesso']

    if ultimo_acesso.month in meses_ferias:
        return 0
    elif dias_sem_acesso > 90:
        return 1
    else:
        return 0
    
df_trn['abandono'] = df_trn.apply(verificar_abandono, axis=1)
df_tst['abandono'] = df_tst.apply(verificar_abandono, axis=1)
```

# removendo colunas não mais nescessarias

```python
colunas_remover = ['dt_primeiro_acesso', 'dt_ultimo_acesso', 'ds_cidade_usuario']
df_trn = df_trn.drop(columns=colunas_remover)
df_tst = df_tst.drop(columns=colunas_remover)
```
